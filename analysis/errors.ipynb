```python
import json
```

```python
with open("../data/errors.json") as file:
    data = json.load(file)
    
    error_types = {
      "dig": 0,
      "ping": 0,
      "traceroute": 0,
    }
    
    for d in data:
      error_types[d["error"]] += 1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(error_types.keys(), error_types.values())
plt.title("Error Count by Call Type")
plt.xlabel("Call Type")
plt.ylabel("Error Count")
```

```python
site_types = {}
for d in data:
  site_type = d["site"].split(".")[1]
  if site_type not in site_types:
    site_types[site_type] = 1
  else:
    site_types[site_type] += 1
```

```python
plt.bar(site_types.keys(), site_types.values())
plt.title("Error Count by TLD Type")
plt.xlabel("TLD Type")
plt.ylabel("Error Count")
```

```python
plt.pie(site_types.values(), labels=site_types.keys())
plt.legend(title="TLD Types", loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1), labels=site_types.keys())
plt.title("Error Count by TLD Type")
```
